```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn.decomposition import PCA
```

```python
df=pd.read_csv("insurance.csv")
df
```

```python
label_encoder = LabelEncoder()
df["smoker"]=label_encoder.fit_transform(df["smoker"])
df["region"]=label_encoder.fit_transform(df["region"])
df["sex"]=label_encoder.fit_transform(df["sex"])
```

```python
df
```

```python
df.info()
```

```python
#arger values of epsilon allow for more errors to be ignored within the margin, potentially leading to a more robust model against noise but with reduced precisio
```

```python
x=df.drop(columns=["charges"])
y=df["charges"]
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
y_scaled = scaler.fit_transform(y.values.reshape(-1, 1))
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y_scaled,test_size=0.2,random_state=343)
model=SVR(kernel="rbf",degree=3,C=2.0,epsilon=0.1)
model.fit(x_train,y_train)
y_pred_test=model.predict(x_test)
y_pred_train=model.predict(x_train)
test_accuracy=r2_score(y_pred_test,y_test)
train_accuracy=r2_score(y_train,y_pred_train)
print("this is test accurcay",test_accuracy)
print("this is train accurcay",train_accuracy)
```

### Applying principal component analysis

```python
pca=PCA(n_components=5)
#if n_components not specified all features it will take or svd_solver="full"
```

```python
principal_components = pca.fit_transform(x_scaled)
principal_components
```

```python
explained_variance = pca.explained_variance_
explained_variance
```

```python
a_train,a_test,b_train,b_test=train_test_split(principal_components,y_scaled,test_size=0.2,random_state=343)
```

```python
a_train.shape
```

```python
model=SVR(kernel="rbf",degree=3,C=2.0,epsilon=0.1)
model.fit(a_train,b_train)
y_pred_test=model.predict(a_test)
y_pred_train=model.predict(a_train)
test_accuracy=r2_score(y_pred_test,b_test)
train_accuracy=r2_score(b_train,y_pred_train)
print("this is test accurcay",test_accuracy)
print("this is train accurcay",train_accuracy)
```
